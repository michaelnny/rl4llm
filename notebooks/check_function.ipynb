```python
import inspect

def logits_processor(param1, param2, **kwargs):
    pass


# Get the function signature
signature = inspect.signature(logits_processor)

# Extract parameters as a list
params = list(signature.parameters.values())

# Check if the third parameter is **kwargs
if len(params) >= 3 and params[2].kind == inspect.Parameter.VAR_KEYWORD:
    print("The third parameter is **kwargs")
else:
    print("The third parameter is NOT **kwargs")

```
