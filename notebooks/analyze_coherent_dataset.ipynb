```python
from collections import Counter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from rl4llm.utils import load_from_jsonl_file
```

```python
sample_files = [
    "../data/synthetic_coherent_dataset/qwen2.5_3b_gsm2k/run_20250319_225709/train.jsonl.gz",
]
```

```python
samples = []
for f in sample_files:
    samples.extend(load_from_jsonl_file(f))

print(f"Total samples: {len(samples)}")
```

```python
# Convert samples to pandas dataframe
df_samples = pd.DataFrame(samples)
```
