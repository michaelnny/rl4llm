```python
import numpy as np
import torch

group_size = 16
min_temp = 0.1
max_temp = 0.9

temperature = torch.linspace(min_temp, max_temp, steps=group_size)

temperature = torch.round(temperature, decimals=2)
print(temperature)
```

```python
group_size = 8
min_temp = 0.1
max_temp = 0.9
desired_step_size = 0.1

# Generate the sequence using the desired step size
steps = torch.arange(0, group_size) * desired_step_size
temperature = min_temp + steps

# Clamp values that exceed max_temp to max_temp
temperature = torch.clamp(temperature, min=min_temp, max=max_temp)
print(temperature)
```

```python
group_size = 64
min_temp = 0.1
max_temp = 0.9
step_sizes=[0.1, 0.05, 0.025, 0.01]

# Calculate the ideal step size based on the range and group size
temp_range = max_temp - min_temp
ideal_step_size = temp_range / (group_size - 1)  # -1 because we include both ends

# Select the largest step size from the list that is <= ideal_step_size
chosen_step_size = max([s for s in step_sizes if s <= ideal_step_size], default=step_sizes[-1])

print(chosen_step_size)
```

```python
# Generate the sequence using the desired step size
steps = torch.arange(0, group_size) * chosen_step_size
temperature = min_temp + steps

# Clamp values that exceed max_temp to max_temp
temperature = torch.clamp(temperature, min=min_temp, max=max_temp)
print(temperature)
```
