```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
from rl4llm.generations import SyntheticDataGenerator
```

```python
generator = SyntheticDataGenerator(model_name="Qwen/Qwen2.5-1.5B-Instruct", output_dir="./test_data/coherence_samples")

# Example 1: Generate from a list of prompts
prompts = [
    "Explain how photosynthesis works.",
    "Write a short story about a detective solving a mystery.",
    "Describe the process of making sourdough bread.",
]
dataset = generator.generate_dataset(prompts, n_samples_per_prompt=16, max_length=1024)

print(len(dataset))
```

```python
incoherent_ds = dataset[dataset['is_coherent'] == False]

print(len(incoherent_ds))
```

```python
for i in range(len(incoherent_ds)):
    row = incoherent_ds.iloc[i]
    print(row['text'])
    print('\n\n--\n\n')
```
