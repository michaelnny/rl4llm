```python
from transformers import AutoTokenizer
```

```python
import os

os.environ['HTTP_PROXY'] = "http://127.0.0.1:1081"
os.environ['HTTPS_PROXY'] = "http://127.0.0.1:1081"
```

```python
model_name = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
full_text = """<think> To solve this problem, we need to first determine

Therefore, Juan needs 600 half-gallons of milk to bake 200 dozen cookies. </think>
<answer> 600 half-gallons </answer>
"""
```

```python
token_ids = tokenizer.encode(full_text)

print(token_ids)
```

```python
for i in range(len(token_ids)):
    token = tokenizer.decode(token_ids[i:i+1])
    print(token)
```
